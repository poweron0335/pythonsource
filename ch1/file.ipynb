```python
# score.txt 를 읽어서 평균과 합 출력

with open("./resource/score.txt", "r") as f:
    list1 = f.readlines()
    hap = 0
    for num in list1:
        # print(num, type(num),end="")
        hap += int(num)
print()
print("점수 총계 : ", hap)
print("평균 : ", hap / len(list1))
```

```python
score = []
with open("./resource/score.txt", "r") as f:
    for num in f:
        # print(num)
        score.append(int(num))

print(sum(score))
print(sum(score) / len(score))
```

```python
# score.txt 읽어서 총합과 평균을 구한 후 result.txt 작성
# result.txt
# 총합 : 520
# 평균 : 86.67

score = []
with open("./resource/score.txt", "r") as f:
    for num in f:
        # print(num)
        score.append(int(num))
        sum1 = sum(score)
        avg = sum1 / len(score)
with open("./resource/result.txt", "w", encoding="utf-8") as result:
    result.write("총합 : %d\n" % sum1)
    result.write("평균 : %.2f\n" % avg)
```

```python
# 500 명의 키와 몸무게를 랜덤하게 생성 후 파일에 저장
import random
strings = list("가나다라마바사아자차카타파하")

print(random.choice(strings) + random.choice(strings)) # 두 자리 이름 생성
print(random.randrange(150, 200)) # 키 만들기
print(random.randrange(50, 100)) # 몸무게 만들기



with open("./resource/info.txt", "w", encoding="utf-8") as info:
    for i in range(1, 500):
            name = random.choice(strings) + random.choice(strings)
            height = random.randrange(150, 200)
            weight = random.randrange(50, 100)
            info.writelines("{}, {}, {}\n".format(name, height, weight))
```

```python
# info.txt 읽어서 BMI 지수 계산

with open("./resource/info.txt", "r", encoding="utf-8") as f:
    for row in f:
        name, height, weight = row.strip().split(", ") # 공백 제거 후 , 를 기준으로 자르기
        print(name, height, weight)
        bmi = int(weight) / ((int(height) / 100) ** 2)
        result = None

        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상체중"
        else:
            result = "저체중"
        # print(f"이름 : {name}\n")
        # print(f"키 : {height}\n")
        # print(f"몸무게 : {weight}\n")
        print("\n".join(["이름: {}", "키 : {}", "몸무게 : {}", "BMI : {}", "결과 : {}\n"]).format(name, height, weight, bmi, result))
```

```python
content = None

while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료 중 원하는 메뉴 번호 입력 : "))

    if no == 1:
        # origin.txt 읽은 후 content 변수에 담기
        # content 변수에 담은 값을 chr(ord(가) + 100) 변경 후 파일에 작성
        # 파일명 : encryption.txt
        with open("./resource/origin.txt", "r", encoding="utf-8") as f:
           content = f.read()
        with open("./resource/encryption.txt", "w", encoding="utf-8") as f:
           for c in content:
               f.write(chr(ord(c) + 100))
    elif no == 2:
        # encryption.txt 읽어온 후 원래의 문자로 출력
         with open("./resource/encryption.txt", "r", encoding="utf-8") as f:
            content = f.read()
            for i in range(0, len(content)):
                print(chr(ord(content[i])-100), end="")
    else:
        break
```
